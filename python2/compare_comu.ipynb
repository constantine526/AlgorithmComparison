```python
import comu as c
from time import time
import matplotlib.pyplot as plt
```

```python
def personnes(amis):
    """
    retourne le tableau des différentes personnes du tableau amis
    """
    personnes=[]
    i=0
    while i<len(amis):
        if amis[i] not in personnes:
            personnes.append(amis[i])
        i+=1
    return personnes

def ses_amis(amis, prenom):
    """
    Retourne le tableau des amis de prenom
    """
    ses_amis=[]
    i=0
    while i<len(amis)/2:
        if amis[2*i]== prenom :
            ses_amis.append(amis[2*i+1])
        elif amis[2*i+1]==prenom :
            ses_amis.append(amis[2*i])
        i+=1
    return ses_amis

def dico_reseau(amis):
    reseau = {}
    pers = personnes(amis)
    i = 0
    while i < len(pers):
        reseau[pers[i]] = ses_amis(amis,pers[i])
        i+=1
    return reseau
```

## Question 2)

### Théorique

Théoriquement, la fonction cree_reseau sera plus efficace que dico_reseau, car elle construit le réseau en un seul passage sur le tableau mis en paramètre. Avec à l'utilisation d'une seule boucle, sa complexité est linéaire, soit O(n).

En revanche, la fonction dico_reseau effectue plusieurs parcours de la liste. D'abord pour rechercher les prénoms à l'aide d'une fonction externe contenant une boucle, puis pour rechercher les amis de chaque prénom, encore une fois via une autre fonction externe utilisant également une boucle. Ainsi, l'algorithme utilise indirectement trois boucles imbriquées, et pour chaque prénom, ce qui nous amène à une complexité de O(p×n), où p est le nombre de prénoms différents. Ce qui va forcament rendre son exécution beaucoup plus lente à mesure que la taille des données (nombre de prénoms) augmente.

### Pratique

Pour la comparaison pratique, nous avons choisi de mesurer le temps d'exécution des deux fonctions en utilisant le même tableau de données. On a multiplier le tableau par 1000 afin de l'agrandir, suffisamment pour pouvoir générer des écarts de temps significatifs, nous permettant ainsi d'observer clairement les différences de performances entre les deux fonctions :

```python
# Générer un grand nombre de relations
large_tab_reseau = ["Alice", "Bob", "Alice", "Dan", "Bob", "Carl", "Dan", "Bob"] * 1000

# Mesure du temps d'exécution de cree_reseau
debut = time()
c.cree_reseau(large_tab_reseau)
print("Temps pour cree_reseau:", round(1000*(time() - debut), 2), "ms")

# Mesure du temps d'exécution de dico_reseau
debut = time()
dico_reseau(large_tab_reseau)
print("Temps pour dico_reseau:", round(1000*(time() - debut), 2), "ms")
```

### Analyse 

La comparaison des temps d'exécution montre que la fonction cree_reseau est environ 5 fois plus rapide que dico_reseau (0.92 ms contre 4.41 ms lorsqu'on l'a lancé de notre côté).
Le parcours unique de la fonction cree_reseau a eu raison de dico_reseau, qui utilisait plutôt des fonctions externes. cree_reseau reste la fonction optimale pour de grandes données.

## Question 11)

### Théorique

La fonction comu_dans_reseau utilise deux fonctions externes, parmi lesquelles tri_popu a la pire complexité, estimée à O(n²). Par conséquent, dans le pire des cas, on va s'attarder sur cette complexité. Cette complexité signifie que son temps d'exécution augmente de manière quadratique avec le nombre total de personnes dans le réseau. Cela s'explique par le fait qu'elle trie d'abord les membres du réseau, puis vérifie les relations d'amitié entre eux, elle fait donc beaucoup de comparaisons.

La fonction comu_dans_amis utilise les mêmes fonctions externes, mais non pas sur les clés du réseau mais sur une valeur (tableau d'amis) donnée. Elle aura donc aussi une complexité de O(m²) où m représente le nombre d'amis d'une personne donnée, elle voit donc son temps d'exécution principalement influencé par le nombre d'amis de la personne mise en paramètre. Cela la rend plus rapide que comu_dans_reseau pour des cas spécifiques où l'on s'intéresse à une seule personne. Cependant, si cette personne possède beaucoup nombre d'amis, l'exécution peut devenir coûteuse à cause du grand nombre de comparaisons qui seront nécessaires pour construire la communauté.

### Pratique

Pour la comparaison, nous avons mesurer les temps d'exécution des deux fonctions sur un même réseau d'amis, en appliquant comu_dans_reseau à l'ensemble du réseau et comu_dans_amis à plusieurs prénoms, afin d'observer les différences de performances en fonction de la taille des données traitées. On a généré un graphique pour que ce soit mieux visible.

```python
dico = c.cree_reseau(large_tab_reseau)

# Liste des prénoms à tester
prenoms = ["Alice", "Bob", "Carl", "Dan"]

# Mesure des temps pour comu_dans_reseau
start = time()
c.comu_dans_reseau(dico)
print("Temps pour comu_dans_reseau:", round((time() - start), 4), "s")

temps_comu_amis = []

# Mesure des temps pour chaque prénom
for prenom in prenoms:
    start = time()
    c.comu_dans_amis(prenom, dico)
    temps_comu_amis.append(round((time() - start), 4))


for i in range(len(prenoms)):
    print("Temps pour comu_dans_amis:", prenoms[i], ":", temps_comu_amis[i], "s")

# Graphique
plt.bar(["comu_dans_reseau"], [temps_comu_reseau], color="g", label="comu_dans_reseau")
plt.bar(prenoms, temps_comu_amis, color="b", label="comu_dans_amis")
plt.xlabel("Fonctions")
plt.ylabel("Temps en secondes")
plt.title("Comparaison des performances des fonctions")
plt.legend()
plt.show()
```

### Analyse

L'analyse des résultats montre que comu_dans_reseau est légèrement plus rapide car elle traite l'ensemble du réseau en une seule fois, tandis que comu_dans_amis voit son temps d'exécution augmenter proportionnellement au nombre d'amis de la personne analysée, rendant son utilisation peu optimale pour des personnes ayant un très grand cercle d'amis.
